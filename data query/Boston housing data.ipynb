# Performance measures LinearRegression
## Mean Squared Error(MSE)

```python
x=df["RM"].values.reshape(-1,1)
y=df["MEDV"].values
```

```python
#Bifraction into train and test ,import test-test-split
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state =20)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
LR = LinearRegression()
```

```python
LR.fit(x_train,y_train)
```

```python
LR.coef_
```

```python
LR.intercept_
```

```python
Yp =LR.predict(x_test) #yp predict
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error,r2_score
from statsmodels.tools.eval_measures import rmse
MAE = mean_absolute_error(y_test,Yp)
MSE= mean_squared_error(y_test,Yp)
r2 =r2_score(y_test,Yp)
RMSE = rmse(y_test,Yp)
print('MAE=',MAE,'\nMSE =',MSE,'\nRMSE=',RMSE,'\nr2=',r2)
```

# Assumption of  Regression Analysis

1. Linear Relationship between the features and target: linear relationship between the features and target

2. Little or no Multicollinearity between the features : Is a state of very high inter-correlations among the independent
   variables.Pair plots and heatmaps(correlation matrix) can be used for identifying highly correlated features.
Polynominal
correlation matrices
## variance inflation factor is 1/1-r2:
Directly proportional to r2 value more the r2 values more will be inflation factor.

low = 0 to 1
moderate = 1 to 5
high >5

r2 value give yuh goodness of best fit line values ranges between 0 to 1
adjustted r squared also gives yuh goodness best fit line considering whether r2 is using all the useful column or not.lesser the adjusted r2 value better will goodness of best fit line.

3. No or low Heteroscedasticity: error term should not explain other error term behaviors, when plotted scatter plot of erro term they should random spread instead showing any pattern.

4. Normal Distribution of error term

5. Little or No autocorrelation in the residuals term
